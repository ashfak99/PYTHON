## 1. Program

```python
def alphaUpperPyramid(size):
    for i in range(0,size+1):
        for j in range(size-i):
            print(' ',end=' ')
        for j in range(2*i+1):
            print(chr(65+j),end=' ')
        print()

alphaUpperPyramid(n)
```
