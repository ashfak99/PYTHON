```python
n=int(input("Please Enter The Size : "))
```

## 1. Program

```python
def alphaUpperPyramid(size):
    for i in range(0,size+1):
        for j in range(size-i):
            print(' ',end=' ')
        for j in range(2*i+1):
            print(chr(65+j),end=' ')
        print()

alphaUpperPyramid(n)
```

## 2. Program

```python
def alphaInvertPyramid(size):
    for i in range(size,-1,-1):
        for j in range(size-i):
            print(' ',end=' ')
        for j in range(2*i+1):
            print(chr(65+i),end=' ')
        print()

alphaInvertPyramid(n)
```

```python
def print_number_pattern(size):
    for i in range(size, -1, -1):
        for j in range(i, size + 1):
            print(j, end="")
        print()
        
    for i in range(1, size + 1):
        for j in range(i, size + 1):
            print(j, end="")
        print()

print_number_pattern(n)
```
