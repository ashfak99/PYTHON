```python
n=int(input("Please Enter The Size : "))
```

## 1.Write angle tringle

```python

def rightTrignle(size):
    for i in range(size):
        for j in range(i+1):
            print("*",end=' ')
        print()


rightTrignle(n)
```

## 2.Decreasing Triangle

```python
def decTrignle(size):
    for i in range(size):
        for j in range(size-i):
            print(' ',end=' ')
        for j in range(i+1):
            print('*',end=' ')
        print()


decTrignle(n)
```

## 3.Inverted Right Half Pyramid

```python
def invtRightHlfPyramid(size):
    for i in range(size):
        for j in range(size-i):
            print("*",end=' ')
        print()

invtRightHlfPyramid(n)
```

## 4.Inverted Left Half Pyramid

```python
def invtLeftHlfPyramid(size):
    for i in range(size):
        for j in range(i):
            print(' ',end=' ')
        for j in range(size-i):
            print('*',end=' ')
        print()

invtLeftHlfPyramid(n)
```

## 5.FULL PYRAMID

```python
def fullPyramid(size):
    for i in range(size):
        for j in range(size-i-1):
            print(" ",end=' ')
        for j in range(2*i+1):
            print('*',end=' ')
        print()


fullPyramid(n)
```

## 6.INVERTED FULL PYRAMID

```python
def invtFullPyramid(size):
    for i in range(size,0,-1):
        for j in range(size-i):
            print(' ',end=' ')
        for j in range(2*i-1):
            print('*',end=' ')
        print()


invtFullPyramid(n)
```

## 7.HALF DIAMOND

```python
def halfDiamond(size):
    for i in range(1, size+1):
        for j in range(i):
            print('*',end=' ')
        print()
        
    for i in range(size-1,0,-1):
        for j in range(i):
            print('*',end=' ')
        print()

halfDiamond(n)
```

## 8.FULL DIAMOND

```python
def diamond(size):
    for i in range(size):
        for j in range(size-i-1):
            print(' ',end=' ')
        for j in range(2*i+1):
            print('*',end=' ')
        print()

    for i in range(size-1,0,-1):
        for j in range(size-i):
            print(' ',end=' ')
        for j in range(2*i-1):
            print('*',end=' ')
        print()


diamond(n)
```
