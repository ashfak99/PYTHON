```python
n=int(input("Please Enter The Size : "))
```

## Write angle tringle

```python

def rightTrignle(size):
    for i in range(size):
        for j in range(i+1):
            print("*",end=' ')
        print()


rightTrignle(n)
```

## Decreasing Triangle

```python
def decTrignle(size):
    for i in range(size):
        for j in range(size-i):
            print(' ',end=' ')
        for j in range(i+1):
            print('*',end=' ')
        print()


decTrignle(n)
```

## Inverted Right Half Pyramid

```python
def invtRightHlfPyramid(size):
    for i in range(size):
        for j in range(size-i):
            print("*",end=' ')
        print()

invtRightHlfPyramid(n)
```

## Inverted Left Half Pyramid

```python
def invtLeftHlfPyramid(size):
    for i in range(size):
        for j in range(i):
            print(' ',end=' ')
        for j in range(size-i):
            print('*',end=' ')
        print()

invtLeftHlfPyramid(n)
```
