# SET

```python
set1 = {1,2,3,4}
set2 = set([3,4,5,6])
```

## Print Set

```python
print("Set 1 : ",set1)
print("Set 2 : ",set2)
```

## Set Functions

```python
set1.add(6)
print("add(6) in set 1 : ",set1)

set1.update([7,8])
print("update([7,8]) in set 1 : ",set1)

copy_set1 = set1.copy()

print("pop() from set 1 : ",set1.pop())

set1.remove(8)
print("remove(8) from set 1 : ",set1)

set1.discard(100)
print("discard(100) set 1: ",set1)

set1.clear()
print("clear() set 1 : ",set1)

A={1,2,3,4}
B={3,4,5,6}

print("Union() : ",A.union(B))
print("Intersection() : ",A.intersection(B))
print("Difference() : ",A.difference(B))
```
