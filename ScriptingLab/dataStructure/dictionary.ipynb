# DICTIONARY

```python
dict1 = {
    'a' : 1,
    'b' : 2,
    'c' : 3
}

dict2 = dict(d=4,e=5,f=6)
```

## Print Dictionary

```python
print("Dict 1 : ",dict1)
print("Dict 2 : ",dict2)
```

## Dictionary Functions

```python
print("len() of dict 1 : ",len(dict1))

print("get('a') from dict 1 : ",dict1.get('a'))

dict1.update({'h':7})
print("update({'h',4}) in dict 1 : ",dict1)

print("keys() of dict 1 : ",dict1.keys())

print("values() of dict 1 : ",dict1.values())

print("items() of dict 1 : ",dict1.items())

print("pop('b) from dict 1 : ",dict1.pop('b'))

print("popitem() dict 1 : ",dict1.popitem())

dict1_copy = dict1.copy()

dict1.clear()
print("clear() dict 1 : ",dict1)
```
