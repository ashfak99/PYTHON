# LIST

```python
list1 = [1,2,3]
list2 = list((4,5,6))
```

## Print List 1 and List 2

```python
print("List 1 : ",list1)
print("List 2 : ",list2)
```

## FUNCTIONS

```python
print("len(list1) : ",len(list1))

list1.append(8)
print("apppend(8 in List 1) : ",list1)

list1.insert(1,10)
print("insert(1,10 in List1) : ",list1)

list1.extend([15,16])
print("extend([15,16]) in List1 : ",list1)

print("Count(4) in List1 : ",list1.count(4))

print("Index(10) in List 1 : ",list1.index(10))

list1.remove(10)
print("remove(10) from List 1 : ",list1)

print("pop() from List1 : ",list1.pop())

list1.reverse()
print("Reverse List1 : ",list1)

list1.sort()
print("sort() List 1 : ",list1)

copy_list = list1.copy()

list1.clear()
print("clear() List 1 : ",list1)
```

## Slicing and Comprehensions

```python
print("List1 copy Slicing : ",copy_list[1:3])

sq=[x**2 for x in range(5)]
print("List Comprehension : ",sq)
```
