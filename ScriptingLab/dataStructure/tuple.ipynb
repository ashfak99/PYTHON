# TUPLE

```python
tuple1 = (1,2,3,2,4)
tuple2 = tuple([5,6,7,8])
```

## Print Tuples

```python
print("Tuple 1 : ",tuple1)
print("Tuple 2 : ",tuple2)
```

## Tuple Functions

```python
print("len(tuple1) : ",len(tuple1))

print("count(2) in tuple 1 : ",tuple1.count(2))

print("index(3) in tuple 1 : ",tuple1.index(3))

print("sorted() tuple 1 : ",sorted(tuple1))

print("min() tuple 1 : ",min(tuple1))

print("max() tuple 1: ",max(tuple1))

print("reversed() tuple 1 : ",tuple(reversed(tuple1)))

print("Comparision (tuple1 == tuple2) : ",tuple1==tuple2)
```
